## Axially Loaded Elastic Bar

Consider the tapered bar

<img src='./Images/strong.d/taperedbar_1.png'/>

Summing forces on the differential volumes gives

$$-p(x) + b\left(x+\frac{\Delta x}{2}\right)\Delta x + p\left(x+\Delta x\right) = 0$$

and taking the limit as $\Delta x \rightarrow 0$ leads to the following differential relationship 

$$\lim_{\Delta x \rightarrow 0} \frac{p\left(x + \Delta x\right) -p\left(x\right)}{\Delta x} + b\left(x+\frac{\Delta x}{2}\right) = \frac{dp}{dx} + b(x) = 0$$

Solution of the differential equation requires a *constitutive equation* relating $p(x)$ to $u(x)$.  Recalling that

$$\sigma(x) = \frac{p(x)}{A(x)} \Rightarrow p(x) = \sigma(x) A(x)$$

and, for a simple linear elastic material, the stress $\sigma$ is given by Hooke's law of linear elasticity<sup>&dagger;</sup>

$$\sigma(x)=E(x)\epsilon(x)$$

where the strain $\epsilon$ is given by

$$\epsilon(x) = \frac{\Delta L}{L} = \lim_{\Delta x\rightarrow 0} \frac{u(\Delta x) - u(x+\Delta x)}{\Delta x} = \frac{du}{dx}$$

Substituting the above expressions in to the differential relationship leads to

$$\frac{d}{dx}\left(A(x)E(x)\frac{du}{dx}\right)+b(x)=0$$

The above is a second-order ordinary differential equation. In the above equation, $u(x)$ is the dependent variable, which is the unknown function, and $x$ is the independent variable. 

To solve the strong form of the governing equation, boundary conditions at each end of the bar must be prescribed.  For the purpose of illustration, we will consider the following specific boundary conditions: at $x=0$, the displacement, $u(x=0)$, is prescribed; at $x=L$, the force per unit area, or traction, denoted by 􏰢$t$, is prescribed. These conditions are written as

$$
\begin{align}
u(x=0) &= \overline{u} \\
\sigma(x=L) &= \left(E\frac{du}{dx}\right)_{x=L}=\frac{F(L)}{A(L)} \equiv \overline{t} 
\end{align}
$$

The traction $􏰢t$ has the same units as stress (force/area), but its sign is positive when it acts in the positive $x$-direction regardless of which face it is acting on, whereas the stress is positive in tension and negative in compression, so that on a negative face a positive stress corresponds to a negative traction.

The differential equation governing the response of the elastic bar along with the associated boundary conditions is called the **strong form** of the governing equation and is summarized here

<div class='msg'>
<p><strong>The strong form</strong></p>
$$
\begin{split}
\frac{d}{dx}\left(A(x)E(x)\frac{du}{dx}\right)+b(x)=0 \\
u(x=0) = \overline{u} \\
\left(E\frac{du}{dx}\right)_{x=L} = \overline{t} 
\end{split}
$$
</div>

**Note**
- The problem data are $\overline{u}$, $\overline{t}$, and $b(x)$ and are given
- Either the load or the displacement can be specified at a boundary point, but not both.

<div class='msg'>
<sup>&dagger;</sup>For thermo-elastic materials, the constitutive model is modified to include thermal effects $\sigma(x) = E(x)\left(\epsilon(x) - \alpha(x)T(x)\right)$
</div>

<div class='msg msg-info'>
<p><strong>Remark</strong></p>
In the previous example there are two types of boundary conditions: one in which the dependent variable ($u$) was specified on the boundary and at the other boundary its derivative was $\left(E du/dx\right)$.  These boundary conditions are classified as <em>Dirichlet</em> and <em>Neummann</em>, respectively.
</div>

## Heat Conduction

Heat flow in a body occurs when there is a temperature difference within the body. Heat transferred in this way is referred to as conduction.  The heat flow through the wall of a heated room in the winter is an example of conduction.  In this section, the strong form of the differential equation governing heat transfer through conduction will be developed.  Other forms of heat transfer such as convection will be discussed in later chapters.

<img src='./Images/strong.d/heat-1.png'/>

Consider a body as shown above. The objective is to determine the temperature distribution. Let $A(x)$ be the area normal to the direction of heat flow and let $s(x)$ be the heat generated per unit thickness of the body, denoted by l. This is often called a heat source. A common example of a heat source is the heat generated in an electric wire due to resistance. In the one-dimensional case, the rate of heat generation is measured in units of energy per time; in SI units, the units of energy are joules (J) per unit length (meters, m) and time (seconds, s). Recall that the unit of power is watts (1 W = 1 J/s). A heat source $s(x)$ is considered positive when heat is generated, i.e. added to the system, and negative when heat is withdrawn from the system. Heat flux, denoted by $q(x)$, is defined as a the rate of heat flow across a surface. Its units are heat rate per unit area; in SI units, W/m$^2$. It is positive when heat flows in the positive x-direction. We will consider a steady-state problem, i.e. a system that is not changing with time.

To establish the differential equation that governs the system, we consider energy balance in a control volume of the wall. Energy balance requires that the rate of heat energy ($qA$) that is generated in the control volume must equal the heat energy leaving the control volume, as the temperature, and hence the energy in the control volume, is constant in a steady-state problem. The heat energy leaving the control volume is the difference between the flow in at on the left-hand side, $qA$, and the flow out on the right-hand side, $q(x + \Delta x)A(x+\Delta x)$. Thus, energy balance for the control volume can be written as

$$s\left(x+\frac{\Delta x}{2}\right)\Delta x + q(x)A(x) -  q\left(x+\Delta x\right)A(x+ \Delta x) = 0$$

Rearranging and taking the limit as $\Delta x \rightarrow 0$ leads to the following differential relationship 

$$\lim_{\Delta x \rightarrow 0} \frac{q\left(x + \Delta x\right)A(x+\Delta x) -q\left(x\right)A(x)}{\Delta x} - s\left(x+\frac{\Delta x}{2}\right) = \frac{d(qA)}{dx} - s = 0$$

The constitutive equation for heat flow, which relates the heat flux to the temperature, is known as Fourier’s law and is given by

$$
q = -k \frac{dT}{dx}
$$

where $T$ is the temperature and $k$ is the thermal conductivity (which must be positive); in SI units, the dimensions of thermal conductivity are W/m$\cdot$<sup>o</sup>C. A negative sign appears because heat flows from high (hot) to low temperature (cold), i.e. opposite to the direction of the gradient of the temperature field.  Substituting Fourier's law gives

$$\frac{d}{dx}\left(Ak\frac{dT}{dx}\right) + s = 0, \quad 0 < x < L$$

Either the flux or the temperature must be prescribed on each end of the bar; these are the boundary conditions. We consider the specific boundary conditions of the prescribed temperature $\overline{T}$ at $x=0$ and prescribed flux $\overline{q}$ at $x=L$. The prescribed flux $\overline{q}$ is positive if heat (energy) flows out of the bar, i.e. $q(x=L)=-\overline{q}$. The strong form for the heat conduction problem is then given by

<div class='msg'>
<p><strong>The strong form</strong></p>
$$\begin{split}
\frac{d}{dx}\left(Ak\frac{dT}{dx}\right) + s = 0, \quad 0 < x < L \\
T=\overline{T} \text{ on } x=0 \\
-q=k\frac{dT}{dx} = \overline{q} \text{ on } x=L\\
\end{split}
$$
</div>

## <a id='eleccond'></a> Electrical Conduction

Consider the tapered bar with electrical potential as shown

<img src='./Images/strong.d/volt-1.png'/>

Conservation of charge requires that

$$i(x) - i\left(x+\Delta x\right) = 0$$

rearranging and taking the limit as $\Delta x \rightarrow 0$ leads to the following differential relationship 

$$\lim_{\Delta x \rightarrow 0} \frac{i\left(x\right) -i\left(x + \Delta x\right)}{\Delta x} = -\frac{di}{dx} = 0$$

Ohm's law gives the constitutive equation relating voltagel and current 

$$\begin{split}
V(x) - V(x+\Delta x) = i \frac{\rho(\Delta x)}{A\left(x+\frac{\Delta x}{2}\right)} \\
-\frac{1}{\rho}A(x)\frac{dV}{dx} = i
\end{split}$$

where $\rho$ is the resistivity.  Substituting the above expressions in to the differential relationship leads to

$$\frac{d}{dx}\left(\frac{1}{\rho}A(x)\frac{dV}{dx}\right)=0$$

which is the strong from of the governing equation.

Boundary conditions take the form of either a prescribed voltage or current at either end.  The strong form for electrical conduction with voltage prescribed at $x=0$ and current prescribed at $x=L$ is given by

<div class='msg'>
<p><strong>The Strong Form</strong></p>
$$
\begin{split}
\frac{d}{dx}\left(\frac{1}{\rho}A(x)\frac{dV}{dx}\right)=0 \\
V(x=0) = \overline{V} \\
\left(\frac{1}{\rho}A\frac{dV}{dx}\right)_{x=L} = \overline{i} 
\end{split}
$$
</div>

<div class='msg msg-info'>
<p><strong>Remark</strong></p>
To find the resistance of the entire element 
<ol>
<li> Prescribe a current BC $i(0) = \overline{i} = 1A$ </li>
<li> Prescribe a voltage BC $V(L) = 0$ </li>
<li> Solve for V(0) </li>
<li> Sove for resistance $R=\frac{V(0) - V(L)}{i(0)} = \frac{V(0)}{1 A}$</li>
</ol>
</div>

## What is being fluxed?

Flux is the rate of "something" per unit area.  To determine what is being fluxed, multiply flux by area to identify the rate of the thing being fluxed

### Heat

The flux, $-qn$,  has units W/m$^2$.  Rate of the “thing” fluxed has units W.
A watt, W, has units of Joules/s.  The thing fluxed has units of Joules.

<div class='boxed w-500'> The thing fluxed is energy! <br>
The heat equation is often called "energy balance"<br>
Steady state: rate of energy is zero!<br>
</div>

### Elasticity

The flux $\bar{t}$ has units N/m$^2$.  Rate of the “thing” fluxed has units N.
A Newton, N, has units kg m/s$^2$. The thing fluxed has units of kg m/s.

<div class='boxed w-500'>
That’s mass times velocity!<br>
The thing fluxed is momentum! <br>
The elasticity equation is often called “momentum balance”<br>
Equilibrium: rate of momentum is zero!<br>
</div>
