## Video Lecture Modules

|[All Lectures](http://physics.oregonstate.edu/~rubin/Books/CPbook/eBook/Lectures/index.html)|[![anything](Figs/RHLlectureMod4.png)](http://physics.oregonstate.edu/~rubin/Books/CPbook/eBook/Lectures/index.html)|
 ![image](Figs/FigB_1a.png)

**Figure B.1** A screenshot from a lecture module showing a dynamic table of
contents, a talking head, video controls and a slide with live
scribbling.

As indicated in the Tables, as an extension
of the concept of a “text”, we provide some 60 lecture modules that
cover most every topic in the book. The modules are a mix of Flash,
Java, HTML and mpeg (or just mpeg for Apples). As seen in the  Figure, each module opens a Web page
containing a video
picture-in-picture of a professor in his office discussing and
demonstrating the material, coordinated dynamic slides (here with red
scribbling on them), a dynamic table of contents, and links to codes and
applets. The modules are based on Greg Moses’s eTeach developments and
use the Camatasia Studio commercial screen capture package. In the face
of rapidly changing technologies, we rely on Camtasia’s regular updates
to keep the modules from becoming obsolete. The modules were directed
and produced by Sally Haerer and took over five years to produce. We
decided that excellent sound quality was essential, as was high
legibility for the slides. The price paid for that is large file sizes,
with the video modules occupying over 14 GB. So while the lectures as
well as the slides are available online, downloading them all at once
may be slow.

The lecture modules can be used to provide another avenue for
understanding the text materials (either as a preview or a review), for
an online course, or in a blended course to replace some lecture time
with lab time (Figure B.2). The ability to vary the degree of
blend to fit the local educational environment and students appears to
significantly improve the effectiveness of a course. This blended
approach, which we recommend, provides time for the instructor to assist
students more personally with their projects and their learning issues.

**Figure B.2** A representation of a blended course in which the instructor chooses
the fraction of face-to-face and all-on-computer time for the different
components of a course.
